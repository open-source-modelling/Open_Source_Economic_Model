```python
def capitalize_unit(policies: pd.DataFrame, market_return: float, is_guaranteed: bool)->pd.DataFrame:
    
    policies["MV"] = (1+market_return)*policies["MV"]

    if is_guaranteed:
        policies["GV"] = (1+market_return)*policies["GV"]
        
    return policies
```

```python
def growth_premium(policies: pd.DataFrame, productivity: float) -> pd.DataFrame:
    
    policies["PREMIUM"] = policies["PREMIUM"] * (1 + productivity)

    return policies
```
